The basketball teams at UConn are the pride and joy of campus. I would like to use my project analyze some potential factors that might help their success in winning another NCAA championship. The men's team often has more fluctuations than the women's team when it comes to personel changes as men only need to stay in college or post high school for one year before they are NBA draft eligible. However, the women's team is much less subject to personnel changes as women are only allowed to be drafted into the WNBA after their junior year. Making the Final Four of the NCAA tournament is often seen as a quintessential achievement for that given season. With almost all of the previous seasons of UConn women's basketball making it to the Final Four, I plan on using this data to predict if the current team will make it to the Final Four.
